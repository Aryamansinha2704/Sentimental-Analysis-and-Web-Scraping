# Sentimental Analysis on WikiPedia Pages of Bernie Sanders and Donald Trump

```python
# import packages
import numpy as np
import pandas as pd
from nltk.sentiment.vader import SentimentIntensityAnalyzer

import requests
import re
from urllib.parse import urlparse
import urllib.robotparser
from bs4 import BeautifulSoup
from datascience import *

# This code checks the robots.txt file
def canFetch(url):

    parsed_uri = urlparse(url)
    domain = '{uri.scheme}://{uri.netloc}/'.format(uri=parsed_uri)

    rp = urllib.robotparser.RobotFileParser()
    rp.set_url(domain + "/robots.txt")
    try:
        rp.read()
        canFetchBool = rp.can_fetch("*", url)
    except:
        canFetchBool = None
    
    return canFetchBool
```

Determining if Bernie Sanders is a democrat or republican by checking the frequency of key words on his wikipedia page

```python
key_words_rep = ["Republican", "republican", "right", "right-wing", "Republican National Convention", "RNC", "rep"]
```

```python
key_words_dem = ["Democrat", "democrat", "left", "left-wing", "Democratic National Convention", "DNC", "dem"]
```

```python
bernie_text = r.text
```

```python
seperating_words = bernie_text.split(" ")
seperating_words
```

```python
def party_finder(text):
    rep_count = 0
    dem_count = 0
    for word in seperating_words:
        for i in key_words_rep:
            if i in word or i.upper() in word or i.lower() in word or i.capitalize() in word:
                rep_count = rep_count + 1
        for i in key_words_dem:
            if i in word or i.upper() in word or i.lower() in word or i.capitalize() in word:
                dem_count = dem_count + 1
    
   
    total = rep_count + dem_count
    dem_percentage = round((dem_count/total) * 100,0)
    rep_percentage = round((rep_count/total) * 100,0)
    diff = abs(dem_percentage)
    print(("Republican Mentions: {0} ({1}% of total)").format(rep_count, rep_percentage))
    print(("Democrat Mentions: {0} ({1}% of total)").format(dem_count, dem_percentage))

   
```

```python
party_finder(bernie_text)
```

As a result, Bernie Sanders appears to be a Democrat which is true.
